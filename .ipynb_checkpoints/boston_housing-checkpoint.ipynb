```python
import pandas as pd
import numpy as np
import requests
import json
from bs4 import BeautifulSoup
```

```python
url = 'http://lib.stat.cmu.edu/datasets/boston'
# Getting the webpage, creating a Response object.
response = requests.get(url)

if response.status_code == 200:
    print('Success!')
elif response.status_code == 404:
    print('Not Found.')

data = response.content
```

```python
soup = BeautifulSoup(data, 'lxml')
```

```python
print(soup.prettify())
```

```python
table = soup.find('p')
print(table.prettify())
```

```python
tokens = table.text.split()
```

```python
for i in range(171, len(tokens)):
    print(tokens[i])
```

```python
boston = []

for i in range(171, len(tokens), 14):
    house = {}
    house['CRIM'] = tokens[i]
    house['ZN'] = tokens[i + 1]
    house['INDUS'] = tokens[i + 2]
    house['CHAS'] = tokens[i + 3]
    house['NOX'] = tokens[i + 4]
    house['RM'] = tokens[i + 5]
    house['AGE'] = tokens[i + 6]
    house['DIS'] = tokens[i + 7]
    house['RAD'] = tokens[i + 8]
    house['TAX'] = tokens[i + 9]
    house['PTRATIO'] = tokens[i + 10]
    house['B'] = tokens[i + 11]
    house['LSTAT'] = tokens[i + 12]
    house['MEDV'] = tokens[i + 13]
    
    boston.append(house)
  

df = pd.DataFrame(boston)
```

```python
df.head()
```

```python
df = df.astype(float)
```

```python
df[['CHAS', 'RAD', 'ZN']] = df[['CHAS', 'RAD', 'ZN']].astype(int)
```
